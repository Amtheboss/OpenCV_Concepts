```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
image1=cv2.imread("/content/drive/MyDrive/Applied AI/ML CB/top_left.jpg")
image2=cv2.imread("/content/drive/MyDrive/Applied AI/ML CB/top_right.jpg")
image3=cv2.imread("/content/drive/MyDrive/Applied AI/ML CB/bottom_left.jpg")
image4=cv2.imread("/content/drive/MyDrive/Applied AI/ML CB/bottom_right.jpg")
image5=cv2.imread("/content/drive/MyDrive/Applied AI/ML CB/center.jpeg")
```

```
# make all the images of same size 
#so we will use resize function
image1=cv2.resize(image1,(200,200))
image2=cv2.resize(image2,(200,200))
image3=cv2.resize(image3,(200,200))
image4=cv2.resize(image4,(200,200))
image5=cv2.resize(image5,(100,100))
```

```
image1_b = cv2.copyMakeBorder(image1, 10, 5, 10, 5, cv2.BORDER_CONSTANT,value=[61, 235, 52])
image2_b = cv2.copyMakeBorder(image2, 10, 5, 5, 10, cv2.BORDER_CONSTANT,value=[61, 235, 52])
image3_b = cv2.copyMakeBorder(image3, 5, 10, 10, 5, cv2.BORDER_CONSTANT,value=[61, 235, 52])
image4_b = cv2.copyMakeBorder(image4, 5, 10, 5, 10, cv2.BORDER_CONSTANT,value=[61, 235, 52])
image5_b = cv2.copyMakeBorder(image5, 5, 5, 5, 5, cv2.BORDER_CONSTANT,value=[61, 235, 52])
```

```
Horizontal1=np.hstack([image1_b,image2_b])
Horizontal2=np.hstack([image3_b,image4_b])
```

```
Vertical_attachment=np.vstack([Horizontal1,Horizontal2])
```

```
plt.imshow(Vertical_attachment)
plt.axis("off")
plt.show()
```

```
Vertical_attachment_rgb = cv2.cvtColor(Vertical_attachment, cv2.COLOR_BGR2RGB)
```

```
plt.imshow(Vertical_attachment_rgb)
plt.axis("off")
plt.show()
```

```
image5_b_rgb = cv2.cvtColor(image5_b,cv2.COLOR_BGR2RGB)
plt.imshow(image5_b_rgb)
plt.axis("off")
plt.show()
```

```
cv2.imwrite('/content/drive/MyDrive/Applied AI/ML CB/background.png',cv2.cvtColor(Vertical_attachment_rgb, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/Applied AI/ML CB/resized.png',cv2.cvtColor(image5_b_rgb, cv2.COLOR_RGB2BGR))
```

```
# load resized image as grayscale
img = cv2.imread('/content/drive/MyDrive/Applied AI/ML CB/resized.png')
h, w, c = img.shape
print(h,w)

# load background image as grayscale
back = cv2.imread('/content/drive/MyDrive/Applied AI/ML CB/background.png')
hh, ww, cc = back.shape
print(hh,ww)

# compute xoff and yoff for placement of upper left corner of resized image   
yoff = round((hh-h)/2)
xoff = round((ww-w)/2)
print(yoff,xoff)

# use numpy indexing to place the resized image in the center of background image
result = back.copy()
result[yoff:yoff+h, xoff:xoff+w,:] = img
```

```
plt.imshow(result)
plt.axis("off")
plt.show()
```

```
result_rgb = cv2.cvtColor(result,cv2.COLOR_BGR2RGB)
```

```
plt.imshow(result_rgb)
plt.axis("off")
plt.show()
```
